```python
# Scatter plot
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,30)
y=np.sin(x)
plt.scatter(x,y)
```

```python
plt.scatter(x,y,c='red',s=50)
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
print(iris.DESCR)
```

```python
features=iris.data.T
plt.scatter(features[0],features[1],s=100*features[3],c=iris.target,alpha=0.5)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

***Histogram plot***

```python
stud_1=np.random.randint(18,45,100)
stud_2=np.random.randint(15,40,100)
plt.hist(stud_1,bins=5)
plt.xlabel('age category')
plt.ylabel(" no of student in that age")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(stud_1,bins,histtype='bar',orientation='vertical',color='c')
plt.show()
```

```python
help(plt.hist)
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(9,5))
plt.hist([stud_1,stud_2],bins,rwidth=0.8,histtype='bar',orientation='vertical',color=['m','y'])
plt.title("histogram example")
plt.xlabel('age category')
plt.ylabel(" no of student in that age")
plt.show()
```

***Pie Plot***

```python
x=[15,25,25,30,5]
plt.pie(x)
plt.show()
```

```python
# adding the labels
labels=['Very likely','likely','Unsure','Unlikely','Very Unlikely']
plt.pie(x,labels=labels)
plt.show()
```

```python
#adding the colors
colors=['red','green','blue','cyan','magenta']
plt.pie(x,labels=labels,colors=colors)
plt.show()
```

```python
plt.pie(x,labels=labels,colors=colors,autopct="%.0f%%")
plt.title("Survey Responses")
plt.show()
```

```python
explode=[0,0,0,0,0.2]
plt.pie(x,labels=labels,colors=colors,autopct="%.0f%%",explode=explode)
plt.title("Survey Responses")
plt.show()
```

```python
explode=[0,0,0,0,0.9]
plt.pie(x,labels=labels,colors=colors,autopct="%.0f%%",explode=explode,shadow=True,startangle=180)
plt.title("Survey Responses")
plt.show()
```

***Stack Plot***


```python
x=[1,2,3,4,5]
y1=[5,6,4,5,7]
y2=[1,6,4,5,6]
y3=[1,1,2,3,2]
plt.stackplot(x,y1,y2,y3)
plt.show()
```

```python
x=[1,2,3,4,5]
y_values={"y1":[5,6,4,5,7],"y2":[1,6,4,5,6],"y3":[1,1,2,3,2]}
plt.stackplot(x,y_values.values())
plt.show()
```

```python
y_values
```

```python
plt.stackplot(x,y_values.values(),labels=y_values.keys())
plt.legend(loc="upper left")
plt.show()
```

***Boxplot***

```python
df=pd.read_csv("winequality-red.csv")
df.head()
```

```python
df.isnull().values.any()
```

```python
fixed_acid=df["fixed acidity"]
free_sulphur=df["free sulfur dioxide"]
total_sulphur=df["total sulfur dioxide"]
plt.boxplot(fixed_acid)
plt.show()
```

```python
plt.boxplot(fixed_acid,vert=0)
plt.show()
```

```python
plt.boxplot([fixed_acid,free_sulphur,total_sulphur])
plt.show()
```

```python
plt.boxplot([fixed_acid,free_sulphur,total_sulphur],notch=True)
plt.show()
```

```python
plt.boxplot([fixed_acid,free_sulphur,total_sulphur],notch=True,patch_artist=True)
plt.show()
```

```python
plt.boxplot([fixed_acid,free_sulphur,total_sulphur],meanline=True,patch_artist=True,showmeans=True)
plt.show()
```
