***Catplot***

```python
import seaborn as sns
tips_df=sns.load_dataset("tips")
tips_df
```

```python
sns.catplot(x="day",y="total_bill",data=tips_df,jitter=False)
```

```python
help(sns.catplot)
```

```python
sns.catplot(x="day",y="total_bill",data=tips_df,kind="swarm")
```

```python
dir(sns)
```

```python
sns.catplot(x="smoker",y="tip",data=tips_df)
```

```python
sns.catplot(x="smoker",y="tip",data=tips_df,order=["No","Yes"])
```

```python
sns.catplot(x="total_bill",y="day",data=tips_df,kind="swarm",hue="time")
```

***Boxplot***

```python
sns.catplot(x="day",y="total_bill",kind="box",data=tips_df,hue="sex")
```

```python
tips_df["weekend"]=tips_df["day"].isin(["Sat","Sun"])
tips_df["weekend"]
```

```python
sns.catplot(x="day",y="total_bill",kind="box",data=tips_df,hue="weekend",dodge=False)
```

***Boxenplot***

```python
diamonds_df=sns.load_dataset("diamonds")
diamonds_df
```

```python
sns.catplot(x="color",y="price",data=diamonds_df,kind="boxen")
```

***Viloin Plot***

```python
sns.catplot(x="total_bill",y="day",hue="sex",data=tips_df,kind="violin",bw_method="silverman")
```

```python
help(sns.violinplot)
```

```python
sns.catplot(x="day",y="total_bill",hue="sex",data=tips_df,kind="violin",bw_method="silverman",split=True)
```

```python
sns.catplot(x="day",y="total_bill",hue="sex",data=tips_df,kind="violin",bw_method="silverman",split=True,inner='stick')
```

```python
a=sns.catplot(x="day",y="total_bill",kind="violin",inner=None,data=tips_df)
sns.swarmplot(x="day",y="total_bill",data=tips_df,ax=a.ax,color='k')
```

***Barplot***

```python
titanic_df=sns.load_dataset("titanic")
titanic_df
```

```python
sns.catplot(x="sex",y="survived",kind="bar",data=titanic_df,hue="class")
```

```python
sns.catplot(x="sex",y="survived",kind="bar",data=titanic_df,hue="embarked")
```

***Countplot***

```python
sns.catplot(x="deck",kind="count",data=titanic_df)
```

```python
sns.catplot(x="who",kind="count",data=titanic_df,hue="class")
```

***Point plot***

```python
sns.catplot(x="age",y="class",kind="point",data=titanic_df)
```

```python
sns.catplot(x="sex",y="survived",kind="point",data=titanic_df,hue='class',palette={"First":'r',"Second":'g',"Third":'b'})
```
