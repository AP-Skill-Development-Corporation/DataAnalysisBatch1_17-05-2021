```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
tips.shape
```

```python
sns.lineplot(x="total_bill",y="tip",data=tips)
```

```python
sns.lineplot(x='size',y='total_bill',data=tips)
```

```python
help(sns.lineplot)
```

```python
sns.lineplot(x="size",y="total_bill",data=tips,hue="sex",palette='hot',legend="brief",markers=["o","<"])
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.set(style="darkgrid")
sns.lineplot(x="size",y="total_bill",data=tips,hue="day",palette='hot',legend="brief",markers=["o","<"])
```

# Distribution plots
- Univariate
- Bivariate 

```python
#univariate
sns.distplot(tips['size'])
```

```python
sns.distplot(tips['tip'])
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['tip'],bins=20)
```

```python
help(sns.distplot)
```

```python
sns.distplot(tips['tip'],hist=False)
```

```python
sns.distplot(tips['tip'],hist=False,kde=False,rug=True)
plt.show()
```

```python
sns.distplot(tips['tip'],color='r',vertical=True)
plt.show()
```

```python
sns.distplot(tips['tip'],color='r',vertical=True,label="tip")
plt.legend()
plt.show()
```

```python
sns.distplot(tips['tip'],hist_kws={'color':'k','edgecolor':'r','alpha':0.9,'linewidth':5,'linestyle':'--'})
plt.legend()
plt.show()
```
