### Creating DataFrame

```python
import pandas as pd
l=[12,90,87,65,45]
df=pd.DataFrame(l)
df
```

```python
#Converting list into Dataframe
l=[[1,2,3],[4,5,6],[9,8,7]]
df1=pd.DataFrame(l)
df1
```

```python
df1.index=["a","b","c"]   #reindexing rows index
df1
```

```python
df1.columns=["A","B","C"]   #reindexing columns
df1
```

```python
#creating dataframe using pandas series

s=pd.Series([79,89,65,45],index=[1,2,3,4])
df=pd.DataFrame(s)
df
```

```python
#creating dataframe using dict

d={"a":90,"b":100,"c":120,"d":140}
df=pd.DataFrame(d,index=[1])
df
```

```python
d={"a":pd.Series([10,20,30,40],index=[1,2,3,4]),
   "b":pd.Series([10,20,30],index=[1,2,3]),
   "c":pd.Series([20,30,45],index=[2,3,5])
  }
df=pd.DataFrame(d)
df
```

### Rows having index values and columns having meaningful name/info nabout data

```python
f=[[i**2 for i in range(10,21)],[j**3 for j in range(10,21)]]
pd.DataFrame(f,index=['squares','cubes'],columns=range(10,21))
```

```python
data=[{"squares":n**2,"cubes":n**3}  for n in range(10,21)]
df1=pd.DataFrame(data,index=range(10,21))
df1
```

```python
d={"a":pd.Series([i**2 for i in range(1,5)],index=range(1,5)),
   "b":pd.Series([i**3 for i in range(1,5)],index=range(1,5))
  }
df=pd.DataFrame(d)
df
```

```python
d1=pd.DataFrame([['a',1],['b',2]],columns=['letter','number'])
d2=pd.DataFrame([['c',1],['d',2]],columns=['letter','number'])
d1.append(d2,ignore_index=True)  
```

```python
d1=pd.DataFrame({"emp":pd.Series(["a","b","c","d"],index=[1,2,3,4]),
                 "yr":pd.Series([2000,2001,2002,2003],index=[1,2,3,4])
                })
print(d1)
d2=pd.DataFrame({"emp":pd.Series(["a","b","c","e"],index=[1,2,3,4]),
                "dept":pd.Series(["hr","op","ma","s"],index=[1,2,3,4])
                })            
print(d2)
```

```python
pd.merge(d1,d2)  #based on emp names
```

```python
pd.merge(d1,d2,how="left")  #left join
```

```python
pd.merge(d1,d2,how="right")   #right join
```

```python
help(pd.merge)
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                        'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                       'value': [5, 6, 7, 8]})
print(df1)
print(df2)
```

```python
df1.merge(df2)
```

```python
df1.merge(df2,how="left")
```

```python
df1.merge(df2,how="right")
```
