```python
import seaborn as sns
df=sns.load_dataset("penguins")
df.head()
```

```python
sns.displot(df,x="bill_length_mm",y="bill_depth_mm")
```

```python
help(sns.displot)
```

```python
sns.displot(df,x="bill_length_mm",y="bill_depth_mm",kind='kde')
```

```python
sns.jointplot(x="bill_length_mm",y="bill_depth_mm",data=df)
```

```python
sns.jointplot(x="bill_length_mm",y="bill_depth_mm",data=df,kind='kde')
```

```python
sns.jointplot(x="bill_length_mm",y="bill_depth_mm",data=df,kind='kde',hue='sex')
```

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
cancer_df=pd.DataFrame(cancer_dataset.data,columns=cancer_dataset.feature_names)
cancer_df["target"]=cancer_dataset.target
cancer_df
```

```python
sns.pairplot(cancer_df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])
```

```python
sns.pairplot(cancer_df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'],hue='target')
```

```python
sns.pairplot(cancer_df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'],hue='target',hue_order=[1,0])
```

```python
sns.pairplot(cancer_df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness'],hue='target',palette="Dark2")
```

```python
sns.pairplot(cancer_df,x_vars=['mean radius','mean texture'],y_vars=['mean perimeter'])
```

```python
sns.pairplot(cancer_df,x_vars=['mean radius','mean texture'],y_vars=['mean perimeter','mean area'],hue="target")
```

```python
sns.pairplot(cancer_df,x_vars=['mean radius','mean texture'],y_vars=['mean perimeter','mean area'],hue="target",markers=['>','^'])
```

```python
titanic_df=sns.load_dataset("titanic")
titanic_df
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue="sex")
```

```python
sns.scatterplot(x=titanic_df.age,y=titanic_df.fare)
```

```python
sns.scatterplot(x=titanic_df["age"],y=titanic_df["fare"])
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue="sex",hue_order=['female','male'])
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,size="who")
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,size="who",sizes=[10,50,80])
```

```python
help(sns.scatterplot)
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,size="who",sizes=[10,50,80],size_order=['child','woman','man'])
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,style="who")
```

```python
sns.scatterplot(x='age',y="fare",data=titanic_df,hue="sex",palette="rainbow")
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,style="who",markers=['3','1','2'],legend=False)
```

```python
import matplotlib.pyplot as plt
kwargs={'edgecolor':'r','facecolor':'k','linewidth':3,'linestyle':'--'}
sns.scatterplot(x="age",y="fare",data=titanic_df,**kwargs,legend="auto")
plt.savefig("Saving the figure")
```
